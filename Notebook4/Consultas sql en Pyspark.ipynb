## obtener el ID de trayecto y contar los ID de pasajeros que hicieron ese trayecto en los años 2017 y 2019.

```python
%sql
SELECT id_trayecto, COUNT(DISTINCT id_pasajero) AS num_personas
FROM viaje
WHERE YEAR(inicio) = 2017 AND MONTH(inicio) = 7
GROUP BY id_trayecto
ORDER BY num_personas DESC;
```

## obtener la fecha de inicio y final donde se hizo más trayectos

```python
%sql
SELECT inicio, fin, COUNT(id_trayecto) AS num_trayectos
FROM viaje
GROUP BY inicio, fin
ORDER BY num_trayectos DESC
LIMIT 1;
```
